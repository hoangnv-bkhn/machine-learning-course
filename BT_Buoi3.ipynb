```
import numpy as np
```

```
from sklearn.datasets import load_iris
data = load_iris()
X, Y = data.data, data.target
```

```
# only classify class 1 vs. class 2
idx = Y > 0
X = X[idx, :]
Y = Y[idx]
Y
```

```
"""import pandas as pd
import numpy as np
import scipy.stats as stats
stats.zscore(X)"""

def z_score(x):
  sdev = np.std(x.T)
  mean = np.mean(x.T)
  z = (x - mean)/sdev
  return z
for i in range(0,4):
  X[:,i] = z_score(X[:,i])
Y = Y-1
```

```
def min_max_scaling(X):
    X_copy = np.copy(X)
    std = np.std(X_copy, 0, float)
    mean = np.mean(X_copy, 0)
    zScore = (X_copy - mean) / std
    return np.concatenate((np.ones((X_copy.shape[0], 1)), zScore), axis = 1)
min_max_scaling(X)
```

```
X = min_max_scaling(X)
weight = np.zeros(X.shape[1])
```

```
def accuracy(X, Y, weight):
    '''
    Evaluate the model, represented by `weight`, with data (X, Y).
    
    Input:
        X:      data features
        Y:      data labels
        weight: model weights
    Ouput:
        Model accuracy on input data.
    '''
    
    # implement your code here
    y = []
    Z = X @ weight
    for x in 1/(1+np.exp(-Z)):
      if x > 0.5:
        y.append(1)
      else:
        y.append(0)
    accur = float(np.sum(Y == y))/(Y.size)
    return accur
```

```
# learning rate
alpha = 1e-2
# epochs
epoch = 10000
```

```
def sigmoid_func(z):
  return 1/(1+np.exp(-z))
def loss_func(X,Y,weight):
  Z = X @ weight
  h = sigmoid_func(Z)
  loss = -np.sum(Y*np.log(h)+(1-Y)*np.log(1-h))/(Y.size)
  return loss
```

```
for _ in range(epoch):
    w = np.dot(X.T, sigmoid_func(X @ weight) - Y)/(Y.size)
    weight -= alpha*w
print(weight)
```

```
accuracy(X, Y, weight)
```
